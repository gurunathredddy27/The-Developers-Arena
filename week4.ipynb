```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

### Creating visualizations for dataset analysis

```python
vk = pd.read_csv("Batsman.csv")
vk.head(3)
```

```python
vk.isnull().sum()
```

### Pie Chart

```python
# we can understand the distribution of the percentage of the dismissal 
plt.figure(figsize=(7,9))
plt.pie(vk['Dismissal'].value_counts(), labels= vk['Dismissal'].value_counts().index, autopct= "%1.2f%%")
```

### Displot

```python
# With this we can understand the disctribution of the score scroed in between 0-175
sns.distplot(vk['RunsScored'], kde= False)
```

### Box plot

```python
# With this we can understand the mean, median and mode of the dataset
sns.boxplot(x = 'RunsScored', data = vk, color = 'orange', showmeans= True)
```

### Count plot

```python
# With this we can understand the number of dismissal in a more visual format 
sns.countplot(x = vk['Dismissal'])
```

### Bar plot 

```python
import numpy as np

plt.figure(figsize = (20,10))
sns.barplot(y= 'RunsScored', x= 'Opposition',estimator=np.sum, data= vk, ci= True )
```

### Point Plot

```python
# this will show more clear way runscored thorugh the years with the help of up and down

sns.pointplot(x= vk['Year'], y = vk['RunsScored'], estimator = 'sum', ci=False)
```

## Client Project: tips dataset

```python
# data loading 

tips = pd.read_csv("tips.csv")  
```

### Histogram

```python
# the distribution of total bill amounts 

plt.figure(figsize=(6,4))
plt.hist(tips["total_bill"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

### Scatter plot

```python
#shows the relationship between total bill and tip amounts. 

plt.figure(figsize=(6,4))
plt.scatter(tips["total_bill"], tips["tip"], color="red", alpha=0.6)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

### Box plot

```python
# shows how total bills vary across different days

plt.figure(figsize=(6,4))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

### Heat Map

```python
# how strongly the numeric features in the dataset are related to each other

plt.figure(figsize=(6,4))
sns.heatmap(tips.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

### Pair plot

```python
#pairplot showing scatter plots and histograms of the features total_bill, tip, and size, separated by gender

sns.pairplot(tips, vars=["total_bill", "tip", "size"], hue="sex")
plt.show()
```

### Volin Plot

```python
#shows the distribution of total_bill values 

sns.violinplot(y=tips['total_bill'])
plt.title("Violin Plot of Total Bill")
plt.ylabel("Total Bill")
plt.show()
```

### Scatter plot

```python
# the relationship between total bill and tip 

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
#how total bills vary across different days. 

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot - Total Bill by Day")
plt.show()
```

```python
# shows how tips vary between genders. 
sns.violinplot(x='sex', y='tip', data=tips)
plt.title("Tip by Gender")
plt.show()
```

### swarm plot

```python
# swarm plot shows how total bills vary across different days

sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day (Swarm Plot)")
plt.show()
```

```python
# total bill values are distributed across different days (rows) and sex (Male/Female) (columns)
g = sns.FacetGrid(tips, col='sex', row='day')
g.map(sns.histplot, 'total_bill')
plt.suptitle("Total Bill Distribution by Sex and Time", y=1.02)
plt.show()
```
