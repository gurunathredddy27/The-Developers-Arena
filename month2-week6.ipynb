- `Normal distribution` is a continuous probability distribution shaped like a bell curve.
- `Poisson distribution` is a discrete distribution used for counting events over time or space in probability

```python
# Normal distribution

norm = np.random.normal(loc=50, scale=10, size=500)
norm
```

```python
# Plot
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(norm, kde=True)
plt.title("Normal Distribution (mu=50, sigma=10)")
```

```python
# Poisson distribution
pois = np.random.poisson(lam=3, size=500)
pois
```

```python
plt.subplot(1,2,2)
sns.histplot(pois, kde=False)
plt.title("Poisson Distribution (lambda=3)")
plt.show()
```

```python
from scipy import stats
```

```python
# one sample t-test

sample = np.random.normal(50, 5, 30)  
t_stat, p_val = stats.ttest_1samp(sample, 50)  
print("t-statistic:", t_stat, "p-value:", p_val)
```

```python
#Two-Sample t-Test

group1 = np.random.normal(55, 5, 30)  
group2 = np.random.normal(50, 5, 30)  
t_stat, p_val = stats.ttest_ind(group1, group2)  
print("t-statistic:", t_stat, "p-value:", p_val)
```

```python
## Chi-Square Test (Categorical Data)

obs = np.array([[30, 10], [20, 40]])  
chi2, p, dof, expected = stats.chi2_contingency(obs)  
print("Chi2:", chi2, "p-value:", p)
```

```python
##Confidence Interval 

sample = np.random.normal(100, 15, 50)  
mean = np.mean(sample)  
conf_int = stats.norm.interval(0.95, loc=mean, scale=stats.sem(sample))  
print("95% Confidence Interval:", conf_int)
```

## Client Project - Perform statistical analysis for comparing two business strategies
- #### Business Scenario: A company tests two marketing strategies (A & B) and wants to know if there’s a significant difference in sales.

```python
# creating business dataset
np.random.seed(42)

# Group A = Email Campaign
email_revenue = np.random.normal(loc=50, scale=10, size=200)

# Group B = Social Media Ads
social_revenue = np.random.normal(loc=55, scale=12, size=200)

# Combine into DataFrame
data = pd.DataFrame({
    'Revenue': np.concatenate([email_revenue, social_revenue]),
    'Strategy': ['Email']*200 + ['Social Media']*200
})
```

#### Visualize distribution of revenue

```python

sns.histplot(data=data, x="Revenue", hue ="Strategy", kde=True, bins=25)
plt.title("Revenue Distribution by Strategy")
plt.show()
```

- Explanation: Social Media has a slightly higher average spread of revenue

```python
# Boxplot comparison
sns.boxplot(x="Strategy", y="Revenue", data=data)
plt.title("Revenue Comparison between Strategies")
plt.show()
```

- Explanation: Social Media shows a higher median with slightly larger spread.

```python
## Mean revenue comparison

mean_values = data.groupby("Strategy")["Revenue"].mean()
mean_values
```

- Explanation: Average revenue is higher for Social Media.

```python
## Variance and standard deviation
var_std = data.groupby("Strategy")["Revenue"].agg(['var','std'])
var_std
```

- Explanation: Social Media strategy has higher variance.

```python
## Confidence interval for Email campaign

ci_email = stats.t.interval(0.95, len(email_revenue)-1, 
                            loc=np.mean(email_revenue), 
                            scale=stats.sem(email_revenue))

print(f" Confidence interval for Email campaign is: {ci_email}")
```

- Explanation: 95% CI for Email revenue mean.

```python
## Confidence interval for Social Media campaign


ci_social = stats.t.interval(0.95, len(social_revenue)-1, 
                             loc=np.mean(social_revenue), 
                             scale=stats.sem(social_revenue))

print(f" Confidence interval for Email campaign is: {ci_social}")
```

- Explanation: 95% CI for Social Media revenue mean.

```python
# Perform independent t-test

t_stat, p_val = stats.ttest_ind(email_revenue, social_revenue)
t_stat, p_val
```

```python
# Interpret results
if p_val < 0.05:
    print("Reject Null Hypothesis: Strategies differ significantly.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference.")
```

```python
# Visualize distributions with violin plot

sns.violinplot(x="Strategy", y="Revenue", data=data)
plt.title("Revenue Distribution by Strategy (Violin Plot)")
plt.show()
```

- Explanation: Social Media shows a wider distribution at higher values.

```python
# Effect size (Cohen’s d)
cohens_d = (np.mean(social_revenue) - np.mean(email_revenue)) / np.sqrt((np.std(social_revenue)**2 + np.std(email_revenue)**2)/2)
cohens_d

# Explanation: Cohen’s d shows practical significance of strategy difference.
```

```python
 # Q Plot for Email campaign (normality check)
stats.probplot(email_revenue, dist="norm", plot=plt)
plt.title("QQ Plot - Email Campaign Revenue")
plt.show()

## QQ Plot for Social Media campaign

stats.probplot(social_revenue, dist="norm", plot=plt)
plt.title("QQ Plot - Social Media Revenue")
plt.show()
```

```python
# Visualization of mean comparison
sns.barplot(x="Strategy", y="Revenue", data=data, ci=95)
plt.title("Mean Revenue with Confidence Interval")
plt.show()

# Explanation: Bar plot shows Social Media has higher mean with CI.
```

```python
# Business interpretation

print("**Business Insight**")
print("On average, Social Media ads generate higher customer revenue.")
print("Statistical tests confirm the difference is significant.")
print("Company should invest more in Social Media marketing.")
```

```python
## Final conclusion

print(" **Project Summary**")
print("1. Social Media strategy shows higher average revenue.")
print("2. 95% confidence intervals confirm the difference.")
print("3. Independent t-test rejects null hypothesis (p < 0.05).")
print("4. Cohen’s d indicates practical significance.")
print("5. Recommendation: Allocate higher budget to Social Media ads.")
```
